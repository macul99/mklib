```python
import sys
#sys.path.append('/home/macul/libraries/MXNet2Caffe')
import pandas as pd
from collections import namedtuple
import numpy as np
import cv2
import pickle
from PIL import Image
import psutil
def process_df(df_rst):
    df_rst['user_id_top0'] = df_rst['user_id'].apply(lambda x: x[0])
    df_rst['score_top0'] = df_rst['score'].apply(lambda x: x[0])
    df_rst['score_top4'] = df_rst['score'].apply(lambda x: x[4])
    df_rst['angle_top0']=np.arccos(df_rst['score_top0'])
    df_rst['angle_top4']=np.arccos(df_rst['score_top4'])
    df_rst['dst_path_top0'] = df_rst['dst_path'].apply(lambda x: x[0])
    df_rst['dst_path_top0_name']=df_rst['dst_path_top0'].apply(lambda x: x.split('/')[-1].split('.')[0])
    df_rst['src_path_name']=df_rst['src_path'].apply(lambda x: x.split('/')[-2])
    return df_rst
def my_sort(x):
    x.sort()
    return x
def add_columns_suffix(df, suffix):
    old_columns = df.columns.tolist()
    new_columns = []
    for col in old_columns:
        if col != 'src_path':
            col = col+suffix
        new_columns.append(col)
    print(new_columns)
    df.columns = new_columns
    return df
def showMultipleImg(img_list):    
    images = [Image.open(name) for name in img_list]
    widths, heights = zip(*(i.size for i in images))

    total_width = sum(widths)
    max_height = max(heights)

    new_im = Image.new('RGB', (total_width, max_height))

    x_offset = 0
    for im in images:
        new_im.paste(im, (x_offset,0))
        x_offset += im.size[0]
    new_im.show()
```

```python
def is_same_person(in_list, ref_list):
    in_list.sort()
    for i in range(len(in_list)):
        for j in range(i+1,len(in_list)):
            if [in_list[i], in_list[j]] not in ref_list:
                return False
    return True
```

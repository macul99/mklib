```python
import cv2
fd='/media/macul/black/spoof_db/iim/toukong_real_jul_2018'
dst_fd='/media/macul/black/spoof_db/toukong_crop25'

imagePaths = list(paths.list_images(fd))
#print(imagePaths)
for i, imgP in enumerate(imagePaths):
    fn = imgP[imgP.rfind('/')+1:]
    im = cv2.imread(imgP)
    h,w,c = im.shape
    if h>w:
        x = int((h-w)/2)
        im = im[x:x+w,:,:]
    elif h<w:
        x = int((w-h)/2)
        im = im[:,x:x+h,:]
    cv2.imwrite(dst_fd+'/'+fn, im)
```
